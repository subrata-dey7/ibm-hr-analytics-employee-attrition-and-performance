```python
# Identifying Outliers

for i in df2.select_dtypes(include='number').columns:
  plt.figure(figsize=(5, 2))
  sns.boxplot(data=df2, x=i)
  plt.title(i)
  plt.show()
```

```python
df3 = df2.copy()    # Copy of df2 assigned to df3. Any changes made to df3 will not affect df2 & vice-versa
df3
```

```python
# Removing Outliers

def wisker(col):
  q1,q3=np.percentile(col,[25,75])
  iqr=q3 - q1

  lw=q1-1.5*iqr
  uw=q3+1.5*iqr
  return lw,uw

for i in ['MonthlyIncome','NumCompaniesWorked','PerformanceRating','StockOptionLevel','TotalWorkingYears',
          'TrainingTimesLastYear','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']:
  lw,uw = wisker(df3[i])
  df3[i]=np.where(df3[i]<lw, lw, df3[i])
  df3[i]=np.where(df3[i]>uw, uw, df3[i])
```

```python
# Again check outliers removed or not

for i in df3.select_dtypes(include='number').columns:
  plt.figure(figsize=(5, 3))
  sns.boxplot(data=df3, x=i)
  plt.title(i)
  plt.show()
```

### **1) Attrition Distribution (Target Variable Analysis)**

```python
#----------  REPRESENTATION USING BAR CHART ---------------

plt.figure(figsize=(7, 5))
# Count number of employees who left vs stayed
df3['Attrition'].value_counts()

# Plot attrition distribution
ax = sns.countplot(x='Attrition',
                   data=df3,
                   palette={'Yes': 'Salmon', 'No': 'lightgreen'})

# Add labels (counts) on each bar
for i in ax.containers:
  ax.bar_label(i)

# Add title to the plot
plt.title("Employee Attrition Distribution")

# Display the plot
plt.show()


#----------  PERCENTAGE REPRESENTATION USING PIE CHART ---------------

# Count attrition values
counts = df3['Attrition'].value_counts()

# Create pie chart
plt.figure(figsize=(6,6))
plt.pie(counts, labels=counts.index,
        autopct='%1.1f%%',  # show percentages
        startangle=60,  # rotate chart for better look
        colors=['lightgreen','salmon'], # custom colors
        explode=(0.09, 0))  # slightly "explode" the first slice

# Add title
plt.title("Employee Attrition Percentage")
plt.show()
```

**CHART INSIGHT:**

The above charts illustrates the distribution of employee attrition.
*   From the bar chart, it's clear that the number of employees who have not attrited ('No') is significantly higher than those who have attrited ('Yes').
*   This indicates an imbalanced dataset where the majority of employees remain with the company, which is a common characteristic in attrition datasets.
*   From the pie plot we can represent in the percentage format that employees who have not **attrited ('No')** is **83.9%** and employees who have **attrited ('Yes')** is **16.1%**

### **14) Correlation Analysis**

```python
# Keep only important numeric columns for attrition analysis
col = [ 'Age', 'Education', 'JobInvolvement', 'WorkLifeBalance',
       'MonthlyIncome','PercentSalaryHike','StockOptionLevel',
        'JobSatisfaction', 'DistanceFromHome']

corr = df4[col].corr()

plt.figure(figsize=(8,5))
sns.heatmap(
    corr,
    cmap="coolwarm",
    annot=True,     # show correlation values
    fmt='.2f',      # format to 2 decimal places
    linewidth=0.5,  # add grid lines between cells
    cbar=True       # show color bar
)
plt.title("Correlation Heatmap")

# Tilt x-axis labels for readability
plt.xticks(rotation=45, ha='right')
plt.show()
```

**CHART INSIGHT:**
*   **Monthly Income and Age:** There is a moderate positive correlation between `MonthlyIncome` and `Age` (approximately 0.50), suggesting that as employees get older, their monthly income tends to increase.
*   **Weak or No Linear Correlations:** For most other pairs of variables displayed, the correlation coefficients are very close to zero. This indicates that there is little to no linear relationship between these variables (e.g., `JobSatisfaction` with `MonthlyIncome`, `PercentSalaryHike` with `Age`, `WorkLifeBalance` with `Education`).
*   **Implications for Attrition:** This chart primarily shows relationships among potential predictor variables. The lack of strong correlations among these variables (except for Age and MonthlyIncome) suggests that multicollinearity might not be a significant issue if these variables were used together in a predictive model, but it also means that many of these factors move independently of each other.
